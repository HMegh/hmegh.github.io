```julia
using Plots, Primes
plotlyjs()
```

```julia
N=1500;
Ω=zeros(Int64,N);
P=primes(N);

for p in P
    q=p;
    while q<=N
        Ω[q:q:end]= Ω[q:q:end] .+ 1;
        q*=p;
    end
end

plot(1:N,Ω,seriestype=:scatter, legend=false,markersize=2,
    yticks=(1:10, [string(i) for i=1:10]))
```

```julia
histogram(Ω,xticks=(0:10, [string(i) for i=0:10]),legend=false)
```

```julia
even=zeros(Int64,N)
odd = zeros(Int64,N);
even[1]=1; #Ω(1) is even (=0)
for i=2:N
    if mod(Ω[i],2)==0
        even[i]=1;
    else
        odd[i]=1;
    end
end

plot(1:N,cumsum(even))
plot!(1:N,cumsum(odd))
```
